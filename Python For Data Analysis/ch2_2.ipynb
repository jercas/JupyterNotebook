# MovieLens 1M 数据集

## 数据读入——处理dat格式数据集

```python
%pylab inline
import pandas as pd
```

```python
# 以unames列表存储用户信息项，随后用pandas.read_table()将dat表中信息读入到DataFrame对象中
unames = ['user_id','gender','age','occupation','zip']
# 调用read_table()函数从数据地址读取数据源；以sep参数指定文件中的分隔符来隔离数据填入DataFrame表格中；以header参数指定表格名；
#                                           以names参数指定每个axis名
users = pd.read_table('D:/PythonWorkSpace/Python for Data Analysis/Python for Data Analysis Source Code/ch02/movielens/users.dat',
                     sep='::',header=None,names=unames)
users
```

```python
# 以rnames列表存储电影评分信息项
rnames = ['user_id','movie_id','rating','timestamp']

ratings = pd.read_table('D:/PythonWorkSpace/Python for Data Analysis/Python for Data Analysis Source Code/ch02/movielens/ratings.dat',
                     sep='::',header=None,names=rnames)
ratings.head(5)
```

```python
# 以mnames列表存储电影信息项
mnames = ['movie_id','title','genres']

movies = pd.read_table('D:/PythonWorkSpace/Python for Data Analysis/Python for Data Analysis Source Code/ch02/movielens/movies.dat',
                     sep='::',header=None,names=mnames)
movies[:5]
```

```python
# 通过pandas.merge()函数进行表/DataFrame的连接合并；该操作会自动识别axis的重叠情况找到合并/连接/外键
data = pd.merge(pd.merge(ratings,users),movies)
data
```

```python
# DataFrame.loc():通过index索引显示一行数据
data.loc[3]
```

```python
# 可延伸显示某几行/index序列的小数据集合
data.loc[[1,2]]
```

```python
# 可延伸显示一个切片的小数据集合
data.loc[:3]
```

```python
# 也可进一步显示，该行某几列的数据
data.loc[3,['rating','gender']]
```

```python
# 但若想类似的单独只读出某column的数据，如此语法就会报错
#　data.loc['rating']
```

```python
# 需依此语法才可，即显示单列也必须带上行index索引/切片才可
data.loc[:5,['rating','age']]
```

```python
# 在Index序列为string类型不便输出时，也可采用DataFrame()直接根据index索引的顺序调出显示
# 注：在index序列为String类型情况下，iloc['a']会直接保存，iloc[]只接受整型输入
data.iloc[3]
```

```python
# 通过DataFrame.pivot_table(),构造数据透视表，重新构造一个判断不同性别用户对同一电影评分的区别
mean_ratings = data.pivot_table('rating',index=['title'],columns='gender',aggfunc='mean')
mean_ratings[:5]
```

```python
# 按电影名title进行分组，以size()统计各个电影的评分数据条数
ratings_by_title = data.groupby('title').size()
ratings_by_title[:10]
```

```python
# 去除评分数据少于250条的电影,将符合要求的电影名存于Index序列中
active_titles = ratings_by_title.index[ratings_by_title.values >= 250]
active_titles
```

```python
# 从之前的mean_ratings中选取符合分析标准(评分数据条数>250)的行
mean_ratings = mean_ratings.loc[active_titles]
mean_ratings[:5]
```

```python
#　对F列降序排列来分析女性最喜欢的电影
#top_female_ratings = mean_ratings.sort_index(by='F' , ascending=False)
top_female_ratings = mean_ratings.sort_values(by='F' , ascending=False)
top_female_ratings[:5]
```

## 计算评分分歧

```python
# 对mean_ratings加上一个存放平均得分之差的列
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
# 对diff差异分列进行排序并存入sorted_by_diff中
sorted_by_diff = mean_ratings.sort_values(by = 'diff')
# 正取前10行，即为分歧最大且女性观众最喜欢的电影
sorted_by_diff[:5]
```

```python
# 对结果反序，再取前五行，即为分歧最大且男性观众最喜欢的电影
sorted_by_diff[::-1][:5]
```

```python
# 若只是想找出分歧最大的电影，不考虑性别因素，则单计算得分数据的方差或标准差即可
ratings_by_title_std = data.groupby('title')['rating'].std()
ratings_by_title_std
```

```python
# 对标准差数据,Seried对象降序排列
ratings_by_title_std.sort_index()
```
